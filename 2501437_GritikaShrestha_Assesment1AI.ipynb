Task 6- Adding HDI category

Create a new column HDI Category that classifies each country into one of the four official
Human Development Index groups. The classification should be based on the HDI value for the
year 2022. Use the following categories and thresholds defined by the United Nations Development
Programme (UNDP):

HDI Category   HDI Range (hdi)

Low            < 0.550

Medium         0.550 – 0.699

High           0.700 – 0.799

Very.          ≥ 0.800

After creating this new column:

• verify that all countries are classified correctly,

• ensure the updated dataframe includes the new category column.

• Save the final dataframe as HDI category added.csv

```python
#This function takes one HDI value and returns its category
def hdi_category(hdi):
    if hdi < 0.550:
        return "Low"
    elif 0.550 <= hdi <= 0.699:
        return "Medium"
    elif 0.700 <= hdi <= 0.799:
        return "High"
    else:
        return "Very High"

#it creates a new column called 'HDI Category'
hdi_2022_df["HDI Category"] = hdi_2022_df["hdi"].apply(hdi_category)
#it checks for missing values
print("Missing HDI Category values:")
print(hdi_2022_df["HDI Category"].isnull().sum())
#it counts how many countries fall into each HDI category
print("\nHDI Category counts:")
print(hdi_2022_df["HDI Category"].value_counts())
#it displays the first few rows
hdi_2022_df[["country", "hdi", "HDI Category"]].head()
#it saves the updated dataframe to a CSV file
hdi_2022_df.to_csv("HDI_category_added.csv", index=False)
```

Task : 3 Outlier Detection

```python
# Creating a function for finding out the measures of dispersion
def detect_outliers(data, column):
    Q1 = data[column].quantile(0.25)
    Q3 = data[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]  # Find rows outside limits
    return outliers
#it finds outliers in the HDI column
outliers_hdi = detect_outliers(HDI_SouthAsia_2022_df, 'hdi')
#it makes a new column to mark HDI outliers
HDI_SouthAsia_2022_df["outliers_hdi"] = HDI_SouthAsia_2022_df['country'].isin(outliers_hdi["country"])
#it prints the HDI outliers
print(f"hdi outliers:", outliers_hdi[["country", "hdi"]])
#it finds outliers in the GNI per capita column
outliers_gni = detect_outliers(HDI_SouthAsia_2022_df, 'gross_inc_percap')
#it makes a new column to mark GNI outliers
HDI_SouthAsia_2022_df["outliers_gni"] = HDI_SouthAsia_2022_df['country'].isin(outliers_gni["country"])
#it prints the GNI outliers
print(f"gni outliers", outliers_gni[["country", "gross_inc_percap"]])
#it combines HDI and GNI outliers into one column
HDI_SouthAsia_2022_df["outliers"] = HDI_SouthAsia_2022_df["outliers_hdi"] | HDI_SouthAsia_2022_df["outliers_gni"]
# it plots scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(data=HDI_SouthAsia_2022_df, x="gross_inc_percap", y="hdi",
                hue="outliers")
plt.title("Outliers in South Asia: HDI vs GNI per Capita")
plt.xlabel("GNI per Capita", fontsize=12)
plt.ylabel("HDI", fontsize=12)
plt.legend(title='Outlier Status', labels=['Normal', 'Outlier'])
plt.tight_layout()
plt.show()
```

Task-7 : Outlier Detection

```python
def detect_outliers(data, column):#creating a function to find outliers using measures of dsipersion
  Q1 = data[column].quantile(0.25)
  Q3 = data[column].quantile(0.75)
  IQR = Q3 - Q1
  lower_bound = Q1 - 1.5 * IQR
  upper_bound = Q3 + 1.5 * IQR
  outliers = data[(data[column] < lower_bound) | (data[column] > upper_bound)]
  return outliers
#it helps find outliers in the hdi column for southAsia
outliers_hdi_sa= detect_outliers(HDI_SouthAsia_2020_2022, 'hdi')
HDI_SouthAsia_2020_2022["outliers_hdi"] = HDI_SouthAsia_2020_2022['country'].isin(outliers_hdi_sa["country"])
print(f"hdi outliers:",outliers_hdi_sa[["country", "hdi"]])
#it helps find outliers in the gross_inc_percap column for south Asia
outliers_gni_sa = detect_outliers(HDI_SouthAsia_2020_2022, 'gross_inc_percap')
HDI_SouthAsia_2020_2022["outliers_gni"] = HDI_SouthAsia_2020_2022['country'].isin(outliers_gni_sa["country"])
print(f"gni outliers",outliers_gni_sa[["country", "gross_inc_percap"]])
#it helps find outliers in the hdi column for middleeast
outliers_hdi_me= detect_outliers(HDI_MiddleEast_2020_2022, 'hdi')
HDI_MiddleEast_2020_2022["outliers_hdi"] = HDI_MiddleEast_2020_2022['country'].isin(outliers_hdi_me["country"])
print(f"hdi outliers",outliers_hdi_me[["country", "hdi"]])
#it helps find outliers in the hdi column for middleeast
outliers_gni_me = detect_outliers(HDI_MiddleEast_2020_2022, 'gross_inc_percap')
HDI_MiddleEast_2020_2022["outliers_gni"] = HDI_MiddleEast_2020_2022['country'].isin(outliers_gni_me["country"])
print(f"gni outliers",outliers_gni_me[["country", "gross_inc_percap"]])
#combining the hdi and gni outliers as whole outliers column
HDI_SouthAsia_2020_2022["outliers"] = HDI_SouthAsia_2020_2022["outliers_hdi"] | HDI_SouthAsia_2020_2022["outliers_gni"]
HDI_MiddleEast_2020_2022["outliers"] = HDI_MiddleEast_2020_2022["outliers_hdi"] | HDI_MiddleEast_2020_2022["outliers_gni"]
#plotting a scatterplot of gni vs hdi
plt.figure(figsize=(10, 6))
sns.scatterplot(data=HDI_SouthAsia_2020_2022, x="gross_inc_percap", y="hdi",
                hue="outliers")
plt.title("Outliers in South Asia: HDI vs GNI per Capita")
plt.xlabel("GNI per Capita", fontsize=12)
plt.ylabel("HDI", fontsize=12)
plt.legend(title='Outlier Status', labels=['Normal', 'Outlier'])
plt.show()
#Gives us a scatter plot for Middle East(GNIvs HDI)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=HDI_MiddleEast_2020_2022, x="gross_inc_percap", y="hdi",
                hue="outliers")
plt.title("Outliers in Middle East: HDI vs GNI per Capita")
plt.xlabel("GNI per Capita", fontsize=12)
plt.ylabel("HDI", fontsize=12)
plt.legend(title='Outlier Status', labels=['Normal', 'Outlier'])
plt.show()
```

It shows uneven development in health,education and economy while most countries have moderate to high HDI and income, this country is behind in both.
